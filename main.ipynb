```python
import numpy
import pandas

csv_path = 'csv/russia_losses_equipment.csv'
column_name = 'APC'
def get_column_as_array():
    return numpy.array(pandas.read_csv(csv_path)[column_name])
array_Of_APC = get_column_as_array()

print('='*25, 'array of APC losses', '='*25, '\n', array_Of_APC)
print('='*25, 'array of daily losses', '='*25, '\n', numpy.diff(array_Of_APC))
print('\nTOP 5:\t', numpy.sort(array_Of_APC)[-5:])
print('\nLosses in summer:\t', array_Of_APC[213] - array_Of_APC[121])
print('\nAverage from 100 till 500 days:\t', numpy.average(array_Of_APC[100:500:1]))
```

```python
import numpy
import pandas
import matplotlib.pyplot as pyplot

csv_path = 'csv/russia_losses_equipment.csv'
column_name = 'APC'
def get_column_as_array():
    return numpy.array(pandas.read_csv(csv_path)[column_name])
array_Of_APC = get_column_as_array()

pyplot.subplots(figsize=(16, 8))
pyplot.plot(numpy.arange(len(array_Of_APC[-200:])), array_Of_APC[-200:], linestyle="dotted", color="red")
pyplot.savefig('dotted_line_plot.png', dpi=100)
pyplot.legend('Losses')
pyplot.title('APC Looses per 200 days')
pyplot.xlabel('Days')
pyplot.ylabel('APC losses')
pyplot.show()
```
